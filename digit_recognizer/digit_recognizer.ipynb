```python
!pip install pandas numpy

import pandas as pd
import numpy as np

train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")

print(train.shape)
print(test.shape)
```

```python
X = train.drop("label", axis=1)
y = train["label"]

X = X / 255.0
test = test / 255.0
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.1, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

log_model = LogisticRegression(max_iter=1000)

log_model.fit(X_train, y_train)

log_pred = log_model.predict(X_val)

print("Logistic Regression Accuracy:", accuracy_score(y_val, log_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_val)

print("Random Forest Accuracy:", accuracy_score(y_val, rf_pred))
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(n_estimators=200),
}
```

```python
log_test_pred = log_model.predict(test)
```

```python
log_submission = pd.DataFrame({
    "ImageId": range(1, len(log_test_pred) + 1),
    "Label": log_test_pred
})

log_submission.to_csv("logistic.csv", index=False)
```

```python
rf_test_pred = rf_model.predict(test)
```

```python
rf_submission = pd.DataFrame({
    "ImageId": range(1, len(rf_test_pred) + 1),
    "Label": rf_test_pred
})

rf_submission.to_csv("randomforest.csv", index=False)
```
